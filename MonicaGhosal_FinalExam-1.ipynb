(a) Combine all these datasets to obtain a dataset at State-Year level, that contains the GDP, surplus, Compensation, and Chain-type quantity indexes for real GDP for each US State and all years starting from 1997 until 2020. Note that each observation must contain the name of the US State, year, and the four values (GDP, surplus, compensation, and Chain-type quantity indexes for real GDP) (10 points).

```python
df_gdp_result=df_gdp.melt(id_vars=['GeoName','Description'], var_name='Year',value_name='GDP')
df_gdp_result

```

```python
df_surplus_result= df_surplus.melt(id_vars='Year', var_name='GeoName',value_name='Surplus')
df_surplus_result
```

```python
df_compensation_updated = df_compensation.pivot_table(index=['Year','GeoName'])
df_compensation_updated
```

```python
df_compensation_updated = df_compensation_updated.rename(columns={'value':'Compensation'})
```

```python
df_compensation_updated = df_compensation_updated.reset_index()
df_compensation_updated
```

```python
all_dfs=[df_gdp_result,df_surplus_result,df_compensation_updated]
```

```python
df_gdp_result['Year']=df_gdp_result['Year'].astype(int)
df_surplus_result['Year']=df_surplus_result['Year'].astype(int)
df_compensation_updated['Year']=df_compensation_updated['Year'].astype(int)
```

```python
merged_df =pd.merge(left=df_gdp_result,
         right=df_surplus_result,
         on=['GeoName','Year'])
merged_df
```

```python
merge_comp = pd.merge(left=merged_df,
                      right=df_compensation_updated, 
                      on=['GeoName','Year'])
merge_comp
```
